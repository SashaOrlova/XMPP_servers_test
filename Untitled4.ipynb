```python
import os
folder = '/Users/aleksandra.orlova/Documents/НИР_весна/prosody/prosody10_simple/'

data = []
minim = 475673465367265736765737657
maxix = -1
cnt = 0
for file in os.listdir(folder):
    if os.path.isfile(folder  + '/' + file) and file != '.DS_Store' and file.startswith('notes'):
        f = open(folder  + '/' + file)
        for line in f:
            if line != '' and len(line) > 15:
                x, y = line.split(' ')
                if x != '' and y != '':
                    cnt += 1
                    minim = min(minim, int(x))
                    data.append([int(x), -int(y)])
                    maxix = max(-int(y), maxix)
                    if (53127169 == -int(y)):
                        print(file)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = []
y = []
cols = []
for cords in data:
    if (cords[1] < 400):
        x.append(cords[0] - minim)
        y.append(cords[1])
        cols.append('black')

plt.scatter(x, y, c=cols, s=1)
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

d = dict()
for x, y in data:
    letter = round(y/15000) * 15000
    if letter not in d:
        d[letter] = 1
    else:
        d[letter] +=1
num = Counter(d)
x = list(num.values())
y = list(num.keys())

x_coordinates = np.arange(len(num.keys()))
plt.bar(x_coordinates,x)
plt.xticks(x_coordinates,y)
plt.show()
```

```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
import collections

t = [y for x,y in data]
dist = np.percentile(t, 95)
print(dist)

d = dict()
for x, y in data:
    if (y > dist):
        letter = round(y/1000) * 1000
        if letter not in d:
            d[letter] = 1
        else:
            d[letter] +=1
d = collections.OrderedDict(sorted(d.items()))
num = Counter(d)
x = [ c / cnt * 100 for c in list(num.values())]
y = list(num.keys())

x_coordinates = np.arange(len(num.keys()))
plt.rcParams["figure.figsize"] = (10,5)
plt.bar(x_coordinates, x)
plt.xlabel('временая задержка в миллисекундах')
plt.ylabel('процент сообщений')
plt.xticks(x_coordinates, y)
plt.show()
```
